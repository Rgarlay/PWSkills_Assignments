#Theory Questions

####1. What is K-Nearest Neighbors (KNN) and how does it work.

KNN or 'K-Nearest Neighbors' is a ML algorithm. It works by making predictions based on the closest data points to a new input. To predict the class or value for a new point, KNN calculates the distance between that point and all points in the training set, then looks at the 'K' nearest neighbors to decide the class for classification or average value for regression.

Because KNN relies on measuring distances, it is sensitive to the scale of the data and usually requires feature scaling. It does not involve an explicit training phase, making it a lazy learner, and all the computation happens during prediction. Choosing the right number of neighbors, K, is crucial since too small a K can cause noisy predictions and too large a K can oversimplify the results.

####2. What is the difference between KNN Classification and KNN Regression.

KNN Classification is used when the target variable has categorical/class labels. For a new data point, it finds the K nearest neighbors and assigns the class that is most common among those neighbors (majority vote).

KNN Regression is used when the target variable is continuous (numerical). Instead of voting, it predicts the value for a new data point by averaging (or sometimes weighting) the target values of the K nearest neighbors.

####3. What is the role of the distance metric in KNN?

The distance metric helps us determine how close or far apart 2 data points are. It helps us in making prediction, as neighbor with lease distance are chosen.

Some ways of finding distances are Euclidean, manhattan distance.

####4. What is the Curse of Dimensionality in KNN?

The Curse of Dimensionality in KNN refers to how the algorithm's performance drops as the number of features (dimensions) increases. In high dimensions, data points become sparse and distances between them become less meaningful, making it hard for KNN to find truly "close" neighbors.

As a result, KNN may struggle to make accurate predictions, require more data, and become slower. To battle this curse, methods like PCA are useful.

####5. How can we choose the best value of K in KNN?

The best value of K in KNN can be chosen using cross-validation. This involves testing different K values and selecting the one that gives the highest accuracy (or lowest error) on validation data.



####6. What are KD Tree and Ball Tree in KNN?



Both ball tree and KD-tree algorithms are implemented in Python libraries like Scikit-learn, giving users powerful tools to optimize nearest-neighbor search.

KD Tree (K-Dimensional Tree) is a binary tree used to organize points in a K-dimensional space. It splits data along different feature axes at each level of the tree to divide the space into nested regions. In KNN, it helps quickly find the nearest neighbors by eliminating large portions of the search space, making it much faster than brute-force search—especially for low-dimensional data.

A Ball Tree is a binary tree that organizes data points into nested hyperspheres (balls). Each node in the tree represents a group of points contained within a ball, defined by a center and radius. During KNN search, Ball Tree can skip entire balls that are too far from the query point, reducing computations. It's especially useful for high-dimensional or non-uniformly distributed data, where KD Tree becomes less efficient.

####7. When should you use KD Tree vs. Ball Tree?

We should prefer KD tree when:

* Data has low dimensions.
* Features are numerical.
* We want faster queries in structured data spaces.

We should preder Ball Tree when:

* Data has high dimensionality.
* Feature distribution is uneven or sparse.

####10. What is PCA (Principal Component Analysis)?

Principal Component Analysis (PCA) is a dimensionality reduction technique that transforms high-dimensional data into a lower-dimensional form while preserving as much variability as possible.
It does this by identifying new axes, called principal components, which are linear combinations of the original features and capture the directions of maximum variance in the data.
These components are uncorrelated and ordered by the amount of variance they explain. PCA is especially useful for simplifying datasets, reducing noise, and improving the performance of machine learning models by removing redundant or correlated features.

####11. How does PCA work?

First, it centers the data by subtracting the mean of each feature. Then, it calculates the covariance matrix to understand how features vary with each other. Next, it computes the eigenvectors and eigenvalues of this matrix. The eigenvectors represent the directions (principal components), and the eigenvalues indicate how much variance each component captures.

PCA then sorts these components by decreasing eigenvalues and selects the top k components to form a new feature space. Finally, it projects the original data onto this lower-dimensional space, effectively reducing dimensionality while retaining the most important information.

####12. What is the geometric intuition behind PCA?

The geometric intuition behind Principal Component Analysis (PCA) is that it rotates the coordinate system of your data to find new axes (directions) that best capture the spread (variance) of the data.

Geometrically, PCA works like projecting the data onto a new coordinate system formed by these principal components. By dropping the components that capture the least variance (usually dimensions with little spread or noise), we reduce dimensionality. This is like shining a light on the data from an angle where the shadow (projection) retains most of the shape and structure, even in fewer dimensions.










####13. 5 What is the difference between Feature Selection and Feature Extraction?

Feature Selection involves choosing a subset of the original features based on some criteria (like correlation, statistical tests, or model performance). It keeps the original features intact, simply removing the less relevant or redundant ones.

Feature Extraction, on the other hand, involves transforming the original features into a new set of features.
These new features are often combinations of the original ones and aim to capture the most relevant information in fewer dimensions. PCA is a classic example—it creates new features (principal components) that are linear combinations of the original variables.

####15. How do you decide the number of components to keep in PCA?


* Each principal component has an associated eigenvalue, which tells you how much variance it captures. You can compute the cumulative explained variance and choose the number of components that retain.

* Use techniques like cross-validation with downstream models to determine how many components give the best performance (e.g., accuracy, RMSE).

####16. Can PCA be used for classification?

PCA is not a classification or regression algorithm. It can certainly help in reducing the deimensions and achieving faster classification in training. However, by itself, it cannot do classification.

####17. What are the limitations of PCA?

* PCA assumes that the data’s structure can be captured by linear combinations of features.

* he new principal components are combinations of original features, which can make them hard to interpret or relate back to the original variables.

* PCA is sensitive to feature scaling. Without proper normalization or standardization, features with larger scales dominate the principal components.

* PCA can be affected by noise in the data, as it tries to capture variance regardless of whether it’s meaningful signal or noise.

####18. How do KNN and PCA complement each other?

KNN and PCA complement each other well because PCA helps prepare the data for KNN by reducing its dimensionality and improving the quality of distance calculations, which KNN relies on.

High-dimensional data can make KNN inefficient and less accurate due to the “curse of dimensionality,” where distances become less meaningful and computational cost increases. By applying PCA first, you reduce the number of features to a smaller set of principal components that capture most of the important variance, removing noise and redundant information.

This makes the distance measurements in KNN more reliable and speeds up the algorithm.

####19. How does KNN handle missing values in a dataset?

KNN itself does not handle missing values directly, so missing data must be addressed before applying KNN.

* We can manually performing imputation on data during pre-processing.

* We can remove error or null value records from our data.


###20. What are the key differences between PCA and Linear Discriminant Analysis (LDA)?

* PCA is an unsupervised technique that finds directions that maximize the overall variance in the data, without considering class labels.
LDA is a supervised technique that aims to find the feature space that maximizes class separability by maximizing the ratio of between-class variance to within-class variance.

* PCA ignores class labels and looks only at the input features.
LDA uses class labels to find projections that best separate different classes.

* PCA can produce as many components as the original number of features.
LDA can reduce the data to at most (number of classes – 1) dimensions, since it focuses on class discrimination.

####21. Train a KNN Classifier on the Iris dataset and print model accuracy.

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score


iris = load_iris()
X, y = iris.data, iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_scaled, y_train)

y_pred = knn.predict(X_test_scaled)
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

####24. Train a KNN Classifier with different values of K and visualize decision boundaried.

```python
import numpy as np
import matplotlib.pyplot as plt


iris = load_iris()
X = iris.data[:, :2]
y = iris.target

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

k_values = [1, 3, 5, 10]
x_min, x_max = X_scaled[:, 0].min() - 1, X_scaled[:, 0].max() + 1
y_min, y_max = X_scaled[:, 1].min() - 1, X_scaled[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                     np.arange(y_min, y_max, 0.02))

fig, axes = plt.subplots(1, len(k_values), figsize=(16, 4))

for i, k in enumerate(k_values):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_scaled, y)
    Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    axes[i].contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.coolwarm)
    scatter = axes[i].scatter(X_scaled[:, 0], X_scaled[:, 1], c=y, cmap=plt.cm.coolwarm, edgecolor='k')
    axes[i].set_title(f'KNN Decision Boundary (k={k})')
    axes[i].set_xlabel('Sepal length (scaled)')
    axes[i].set_ylabel('Sepal width (scaled)')

plt.tight_layout()
plt.show()
```

####33. Implement KNN Imputation for handling missing values in a dataset.

####38. Train a PCA model and analyze the effect of different numbers of components on accuracy.

```python
iris = load_iris()
X, y = iris.data, iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

accuracies = []
components_range = range(1, X.shape[1] + 1)

for n_components in components_range:
    pca = PCA(n_components=n_components)
    X_train_pca = pca.fit_transform(X_train_scaled)
    X_test_pca = pca.transform(X_test_scaled)

    knn = KNeighborsClassifier(n_neighbors=5)
    knn.fit(X_train_pca, y_train)
    y_pred = knn.predict(X_test_pca)

    acc = accuracy_score(y_test, y_pred)
    accuracies.append(acc)
    print(f"Components: {n_components}, Accuracy: {acc:.4f}")

plt.figure(figsize=(8, 5))
plt.plot(components_range, accuracies, marker='o', linestyle='--')
plt.title('Effect of PCA Components on KNN Accuracy')
plt.xlabel('Number of PCA Components')
plt.ylabel('Accuracy')
plt.grid(True)
plt.xticks(components_range)
plt.tight_layout()
plt.show()
```

####46. Train a PCA model and visualize the data reconstruction error after reducing dimensions.

```python
iris = load_iris()
X = iris.data

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

n_components = range(1, X.shape[1] + 1)

reconstruction_errors = []

for n in n_components:
    pca = PCA(n_components=n)
    X_reduced = pca.fit_transform(X_scaled)
    X_reconstructed = pca.inverse_transform(X_reduced)
    mse = mean_squared_error(X_scaled, X_reconstructed)
    reconstruction_errors.append(mse)


plt.figure(figsize=(8,5))
plt.plot(n_components, reconstruction_errors, marker='o', color='red')
plt.xlabel('Number of Principal Components')
plt.ylabel('Reconstruction Error (MSE)')
plt.title('PCA Reconstruction Error vs Number of Components')
plt.xticks(n_components)
plt.grid(True)
plt.show()
```

####48. Train a PCA model and analyze the effect of different numbers of components on data variance.

```python
data = load_iris()
X = data.data

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

max_components = X.shape[1]

cumulative_variances = []

for n in range(1, max_components + 1):
    pca = PCA(n_components=n)
    pca.fit(X_scaled)
    cumulative_variance = np.sum(pca.explained_variance_ratio_)
    cumulative_variances.append(cumulative_variance)

plt.figure(figsize=(8,5))
plt.plot(range(1, max_components + 1), cumulative_variances, marker='o', linestyle='--', color='b')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance Ratio')
plt.title('Effect of Number of PCA Components on Explained Variance')
plt.xticks(range(1, max_components + 1))
plt.grid(True)
plt.show()
```
