####2. The difference between Deep copy and shallow copy.

1. Shallow copy creates a new object but references the same data.

Deep copy creates a new object and copies all data recursively.

2. Changes to the original data affect the shallow copy’s content.

Changes to the original data do not affect the deep copy’s content.

3. Shallow copy is faster and uses less memory.

Deep copy is slower but safer for independent manipulation.

####14. Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following tasks using Matplotlib:

a) Plot a histogram of the data with 30 bins.

b) Overlay a line plot representing the normal distribution's probability density function (PDF).

c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.

d) Set the title of the plot as 'Histogram with PDF Overlay'.

```python
from scipy.stats import norm

data = np.random.normal(loc=0, scale=1, size=1000)

count, bins, ignored = plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label='Histogram')

xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, loc=np.mean(data), scale=np.std(data))

plt.plot(x, p, 'r-', label='PDF')
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')
plt.title('Histogram with PDF Overlay')
plt.legend()
plt.show()
```

####15. Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.

```python
x = np.random.randn(100)
y = np.random.randn(100)

# Determine quadrant for each point
def quadrant(x, y):
    if x >= 0 and y >= 0:
        return 'Q1'
    elif x < 0 and y >= 0:
        return 'Q2'
    elif x < 0 and y < 0:
        return 'Q3'
    else:
        return 'Q4'

quadrants = [quadrant(xi, yi) for xi, yi in zip(x, y)]

# Create DataFrame
df = pd.DataFrame({'x': x, 'y': y, 'Quadrant': quadrants})

# Plot
sns.scatterplot(data=df, x='x', y='y', hue='Quadrant', palette='Set1')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')
plt.legend(title='Quadrants')
plt.show()
```

```python
from bokeh.plotting import figure, show, output_notebook
import numpy as np

output_notebook()

x = np.linspace(0, 4*np.pi, 100)
y = np.sin(x)

p = figure(title='Sine Wave Function', x_axis_label='x', y_axis_label='y')
p.line(x, y, line_width=2)
p.xgrid.grid_line_color = 'navy'
p.xgrid.grid_line_alpha = 0.5
p.ygrid.grid_line_color = 'navy'
p.ygrid.grid_line_alpha = 0.5

show(p)
```

####17. Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical Bar Chart'.

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.transform import factor_cmap
from bokeh.palettes import Viridis256

output_notebook()

# Generate random categorical data
categories = [f'Category {i}' for i in range(1, 11)]
values = np.random.randint(10, 100, size=len(categories))

# Create DataFrame and ColumnDataSource
df = pd.DataFrame({'category': categories, 'value': values})
source = ColumnDataSource(df)

# Create Bokeh figure with correct size parameters
p = figure(
    x_range=categories,
    height=400,
    width=800,
    title="Random Categorical Bar Chart",
    x_axis_label="Category",
    y_axis_label="Value",
    toolbar_location=None,
    tools=""
)

# Color map for bars
mapper = factor_cmap('category', palette=Viridis256, factors=categories)

# Add vertical bars
p.vbar(x='category', top='value', width=0.8, source=source, fill_color=mapper, line_color='black')

# Add hover tooltips
hover = HoverTool(tooltips=[("Category", "@category"), ("Value", "@value")])
p.add_tools(hover)

# Show the plot in notebook
show(p)
```

####18. Using Plotly, create a basic line plot of a randomly generated dataset, label the axes, and set the title as 'Simple Line Plot'.

```python
import plotly.graph_objects as go
import numpy as np

# Generate random dataset
x = np.arange(0, 50)
y = np.random.rand(50)

# Create the line plot
fig = go.Figure()

fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name='Random Line'))

# Add title and labels
fig.update_layout(
    title='Simple Line Plot',
    xaxis_title='X-axis',
    yaxis_title='Y-axis'
)

# Show the plot
fig.show()
```
