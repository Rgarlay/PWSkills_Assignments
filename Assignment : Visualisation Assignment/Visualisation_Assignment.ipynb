##Bokeh Portion

####1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
```

```python
from bokeh.plotting import figure,output_file, show
```

```python
a = np.arange(0,10,0.1)
x = np.sin(a)
p = figure(title = 'sin_wave')
p.xaxis.axis_label = 'x'
p.yaxis.axis_label = 'sin(x)'
p.line(a,x)
show(p)
```

####2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.

```python
np.random.seed(42)
x = np.random.randint(1,50,size = 25)
y = np.random.randint(1,50,size = 25)

p_1 = figure(title='scatter_plot')
p_1.xaxis.axis_label = 'x_coordinate'
p_1.yaxis.axis_label = 'y_coordinate'
p_1.scatter(x,y,size = 5)
show(p_1)
```

####3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

```python
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

p_2 = figure(x_range = fruits,title = 'Categories')
p_2.xaxis.axis_label = 'Fruits'
p_2.yaxis.axis_label = 'Counts'
p_2.vbar(x=fruits, top=counts,width=0.9)
show(p_2)
```
